# MA: M365 Authentication (INV-CAIO-M365-003, Eq. 4)

**Lemma:** [LEMMA_L3.md](../docs/contracts/caio-m365/LEMMA_L3.md) — Key required and missing/invalid ⇒ status 401.

**Math:** \(\mathtt{keyMissingOrInvalid}(H) \Rightarrow \mathtt{status}(S) = 401\).

**Producer:** `scripts/ci/verify_m365_auth.py` → `configs/generated/m365_auth_verification.json`.

```python
import subprocess
import sys
from pathlib import Path

repo_root = Path.cwd().resolve()
while (
    not (repo_root / "scripts" / "ci" / "verify_m365_auth.py").exists()
    and repo_root != repo_root.parent
):
    repo_root = repo_root.parent
r = subprocess.run(
    [sys.executable, str(repo_root / "scripts" / "ci" / "verify_m365_auth.py")],
    cwd=str(repo_root),
    capture_output=True,
    text=True,
)
print(r.stdout or "")
assert r.returncode == 0
```

```python
import json

with open(repo_root / "configs" / "generated" / "m365_auth_verification.json") as f:
    a = json.load(f)
print("auth_pass:", a.get("auth_pass"))
```
