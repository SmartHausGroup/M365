# Purpose

Prove customer onboarding v2 contract is bounded and deterministic.

## Lemma: L69

```python
from pathlib import Path
import yaml

repo_root = Path.cwd()
contract = yaml.safe_load((repo_root / 'registry' / 'customer_onboarding_v2.yaml').read_text(encoding='utf-8'))
registry = yaml.safe_load((repo_root / 'registry' / 'persona_registry_v2.yaml').read_text(encoding='utf-8'))

assert contract['contract']['id'] == 'customer-onboarding-v2'
assert len(contract['onboarding_phases']) == 4
assert len(contract['governance_rules']) == 5
for pid in contract['kpis']['core_tier_activation_sequence']:
    assert pid in registry['personas']
    assert registry['personas'][pid]['coverage_status'] == 'registry-backed'
```
