# Purpose

Prove the persona discovery contract is bounded, deterministic, and covers all personas.

## Lemma Mapping

- L59 — M365 Persona Discovery Contract v1

## Expected Results

- Six discovery dimensions are declared.
- Selection rules handle all match scenarios.
- Persona count matches the authoritative registry.

```python
from pathlib import Path
import yaml

repo_root = Path.cwd()
contract = yaml.safe_load((repo_root / 'registry' / 'persona_discovery_contract_v1.yaml').read_text(encoding='utf-8'))
registry = yaml.safe_load((repo_root / 'registry' / 'persona_registry_v2.yaml').read_text(encoding='utf-8'))

assert len(contract['discovery_dimensions']) == 6
assert contract['selection_rules']['ambiguity_policy'] == 'fail_closed'
assert contract['kpis']['discoverable_persona_count'] == len(registry['personas'])
assert contract['kpis']['department_count'] == 10
```
