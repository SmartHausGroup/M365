## Runtime Projection

Build the HR department pack through the shared runtime and assert the deterministic summary.


```python
from pathlib import Path
from tempfile import TemporaryDirectory

import yaml

from smarthaus_common.department_pack import build_department_pack
from smarthaus_common.json_store import JsonStore

repo_root = Path.cwd()
authority_path = repo_root / 'registry' / 'department_pack_hr_v1.yaml'
payload = yaml.safe_load(authority_path.read_text(encoding='utf-8'))

assert payload['department']['id'] == 'hr'
assert payload['kpis']['required_personas'] == 1
assert payload['kpis']['supported_action_count'] == 5
assert list(payload['personas']) == ['hr-generalist']
```

```python
with TemporaryDirectory() as temp_dir:
    pack = build_department_pack('hr', store=JsonStore(Path(temp_dir)))

assert pack['department']['id'] == 'hr'
assert pack['summary']['persona_count'] == 1
assert pack['summary']['active_persona_count'] == 1
assert pack['summary']['supported_action_count'] == 5
assert pack['summary']['pack_state'] == 'ready'
assert [persona['persona_id'] for persona in pack['personas']] == ['hr-generalist']
assert pack['personas'][0]['approval_profile'] == 'critical-regulated'
```
