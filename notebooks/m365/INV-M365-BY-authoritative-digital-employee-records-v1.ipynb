# INV-M365-BY — Authoritative Digital Employee Records v1

**Purpose:** certify the bounded employee-record authority for the 20 promoted personas before any authoritative registry rebase.

**Lemma mapping:** `L77`

**Invariant support:** `invariants/lemmas/L77_m365_authoritative_digital_employee_records_v1.yaml`

**Expected result:** `20` complete employee records, `7` covered departments, deterministic manager/escalation tokens, and exactly three bounded humanization fields.


Load the employee-record authority artifact and derive the per-department distribution.


```python
from collections import Counter
from pathlib import Path
import yaml

repo = Path.cwd()
payload = yaml.safe_load((repo / 'registry' / 'authoritative_digital_employee_records_v1.yaml').read_text())
records = payload['records']
required_fields = set(payload['required_fields'])
bounded_fields = set(payload['bounded_humanization_fields'])
departments = Counter(record['department'] for record in records.values())
names = [record['display_name'] for record in records.values()]
slugs = [record['slug'] for record in records.values()]
```

Assert the bounded H2 contract.


```python
assert len(records) == 20
assert payload['summary']['total_promoted_personas'] == 20
assert payload['summary']['departments_covered'] == 7
assert required_fields == {
    'canonical_agent', 'display_name', 'slug', 'department', 'title',
    'manager', 'escalation_owner', 'working_style', 'communication_style', 'decision_style'
}
assert bounded_fields == {'working_style', 'communication_style', 'decision_style'}
assert len(names) == len(set(names))
assert len(slugs) == len(set(slugs))
for agent_id, record in records.items():
    assert set(record.keys()) == required_fields
    assert record['canonical_agent'] == agent_id
    assert record['manager'] == f"department-lead:{record['department']}"
    assert record['escalation_owner'] == f"department-owner:{record['department']}"
```

Produce the deterministic verification preview that H2 uses before registry extraction phases begin.


```python
verification_preview = {
    'total_promoted_personas': len(records),
    'departments': dict(sorted(departments.items())),
    'bounded_humanization_fields': sorted(bounded_fields),
}
verification_preview
```
