# INV-M365-BX — Activated Persona Surface v1

**Purpose:** certify the current activated M365-backed persona surface on `feature/m365_personas`.

**Lemma mapping:** `L76`

**Invariant support:** `invariants/lemmas/L76_m365_activated_persona_surface_v1.yaml`

**Expected result:** `34` registry-backed personas, `5` deferred external personas, `298` total allowed persona-actions, and no commercialization claim that includes deferred external personas in the active surface.


Load the authoritative registries and compute the active/deferred partition directly from repo state.


```python
from pathlib import Path
import yaml

repo = Path.cwd()
persona_registry = yaml.safe_load((repo / 'registry' / 'persona_registry_v2.yaml').read_text())
capability_map = yaml.safe_load((repo / 'registry' / 'persona_capability_map.yaml').read_text())
personas = persona_registry['personas']
active = {k: v for k, v in personas.items() if v['coverage_status'] == 'registry-backed'}
deferred = {k: v for k, v in personas.items() if v['coverage_status'] == 'persona-contract-only'}
active_total_actions = sum(v['action_count'] for v in active.values())
active_departments = {v['department'] for v in active.values()}
sorted(deferred)
```

Assert the certified branch-state boundary.


```python
assert len(active) == 34
assert len(deferred) == 5
assert sorted(deferred) == [
    'app-store-optimizer',
    'instagram-curator',
    'reddit-community-builder',
    'tiktok-strategist',
    'twitter-engager',
]
assert active_total_actions == 298
assert len(active_departments) == 10
assert persona_registry['summary']['registry_backed_personas'] == 34
assert persona_registry['summary']['persona_contract_only_personas'] == 5
assert capability_map['summary']['current_registry_backed_personas'] == 34
assert capability_map['summary']['persona_contract_only_personas'] == 5
```

Verify prior wave scorecards remain green so the aggregated certification claim is bounded by proven wave evidence.


```python
import json

for lemma_id in ['l73', 'l74', 'l75']:
    scorecard = json.loads((repo / 'artifacts' / 'scorecards' / f'scorecard_{lemma_id}.json').read_text())
    assert scorecard['status'] == 'green'
```
