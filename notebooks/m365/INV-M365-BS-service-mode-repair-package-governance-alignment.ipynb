# INV-M365-BS Service-Mode Repair Package Governance Alignment

## Purpose
Prove that the current `m365-token-provider-runtime-repair` child package is not yet aligned to the current master plan/prompt structure, so the planned normalization is grounded in notebook evidence rather than guesswork.

## Scope
- Read the current child plan YAML and prompt pair
- Compare them to the current cross-repo master structure signals
- Record whether normalization is required

## Expected Result
The baseline package should fail current-template alignment checks and therefore require normalization before the repair phase is re-run.

## Imports and Paths
Load the current M365 child package and the current UCP master governance references that define the target structure.

```python
from __future__ import annotations

import json
from pathlib import Path

import yaml

repo_root = Path.cwd().resolve()
plan_yaml_path = repo_root / 'plans/m365-token-provider-runtime-repair/m365-token-provider-runtime-repair.yaml'
detail_prompt_path = repo_root / 'docs/prompts/codex-m365-token-provider-runtime-repair.md'
kickoff_prompt_path = repo_root / 'docs/prompts/codex-m365-token-provider-runtime-repair-prompt.txt'
ucp_root = repo_root.parent / 'UCP'
ucp_schema_path = ucp_root / 'agent_governance/PLAN_TEMPLATE_SCHEMA.json'
ucp_prompt_template_path = ucp_root / 'agent_governance/master-maths-prompt-template.md'

assert plan_yaml_path.exists(), plan_yaml_path
assert detail_prompt_path.exists(), detail_prompt_path
assert kickoff_prompt_path.exists(), kickoff_prompt_path
assert ucp_schema_path.exists(), ucp_schema_path
assert ucp_prompt_template_path.exists(), ucp_prompt_template_path
```

## Baseline Alignment Check
Check for the minimum fields and sections the current master structure requires. This is not an execution notebook for the repair itself; it is governance evidence that the package needs normalization.

```python
plan_yaml = yaml.safe_load(plan_yaml_path.read_text(encoding='utf-8'))
schema = json.loads(ucp_schema_path.read_text(encoding='utf-8'))
detail_prompt = detail_prompt_path.read_text(encoding='utf-8')
kickoff_prompt = kickoff_prompt_path.read_text(encoding='utf-8')
template_text = ucp_prompt_template_path.read_text(encoding='utf-8')

required_plan_keys = {
    'plan_id',
    'parent_plan_id',
    'title',
    'version',
    'status',
    'owner',
    'date_created',
    'date_updated',
    'north_star_ref',
    'execution_plan_ref',
    'domain',
    'math_algorithm_scope',
    'north_star_alignment',
    'intent_capture',
    'objective',
    'current_state',
    'target_state',
    'scope',
    'requirements',
    'execution_sequence',
    'tasks',
    'gates',
    'artifacts',
    'implementation_approach',
    'risks',
    'rollback',
    'traceability',
    'governance_closure',
    'agent_constraints',
    'references',
}

missing_plan_keys = sorted(required_plan_keys - set(plan_yaml))
missing_detail_markers = [
    marker
    for marker in [
        'Execution Prompt',
        'Plan Reference:',
        'Parent Plan:',
        'North Star:',
        'Governance Lock',
        'M - Model',
        'A - Annotate',
        'T - Tie',
        'H - Harness',
        'S - Stress-test',
    ]
    if marker not in detail_prompt
]
missing_kickoff_markers = [
    marker
    for marker in [
        'Read and execute the detailed',
        'Plan Reference:',
    ]
    if marker not in kickoff_prompt
]

result = {
    'schema_title': schema.get('title'),
    'template_header_present': 'Execution Prompt — MATHS Workstream' in template_text,
    'missing_plan_keys': missing_plan_keys,
    'missing_detail_markers': missing_detail_markers,
    'missing_kickoff_markers': missing_kickoff_markers,
    'requires_normalization': bool(
        missing_plan_keys or missing_detail_markers or missing_kickoff_markers
    ),
}
result
```

## Deterministic Assertion
The baseline package must require normalization. If this assertion fails, the premise for the governance rewrite is wrong.

```python
assert result['requires_normalization'] is True
assert result['missing_plan_keys'], 'expected missing full-template plan keys'
assert result['missing_detail_markers'], 'expected missing current prompt-template markers'
print('baseline governance alignment gap proven')
```
