# Purpose

Prove the cross-department workflow certification contract is bounded, deterministic, and complete.

## Lemma Mapping

- L66 — M365 Cross-Department Workflow Certification v1

```python
from pathlib import Path
import yaml

repo_root = Path.cwd()
contract = yaml.safe_load(
    (repo_root / 'registry' / 'cross_department_certification_v1.yaml').read_text(encoding='utf-8')
)
collab = yaml.safe_load(
    (repo_root / 'registry' / 'cross_persona_collaboration_contract_v1.yaml').read_text(encoding='utf-8')
)
dept_cert = yaml.safe_load(
    (repo_root / 'registry' / 'department_certification_v1.yaml').read_text(encoding='utf-8')
)

assert contract['contract']['id'] == 'cross-department-certification'
assert len(contract['certification_phases']) == 4
assert set(contract['governance_rules'].keys()) == {
    'fail_closed', 'audit_completeness', 'bounded_claims',
    'determinism', 'prerequisite_alignment',
}

kpis = contract['kpis']
assert kpis['collaboration_primitive_count'] == len(collab['collaboration_primitives'])
assert kpis['handoff_rule_count'] == len(collab['handoff_rules'])
assert kpis['certified_department_count'] == len(dept_cert['department_certification_status'])
assert contract['cross_department_certification_summary']['overall_verdict'] == 'certified'
```
