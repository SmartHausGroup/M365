# Purpose

Prove the Project Management department pack authority is bounded, deterministic, and fail-closed while the department remains contract-only.

## Lemma Mapping

- L54 — M365 Project Management Department Pack v1

## Invariant Support

- invariants/lemmas/L54_m365_project_management_department_pack_v1.yaml

## Expected Results

- The Project Management pack authority validates.
- The shared runtime projects three contract-only personas.
- The pack deterministically resolves to `blocked` with zero supported actions.


## Authority Validation

Validate the Project Management authority shape and zero-action KPI surface.

## Runtime Projection

Build the Project Management department pack through the shared runtime and assert the deterministic blocked summary.

```
from pathlib import Path
from tempfile import TemporaryDirectory

import yaml

from smarthaus_common.department_pack import build_department_pack
from smarthaus_common.json_store import JsonStore

repo_root = Path.cwd()
authority_path = repo_root / 'registry' / 'department_pack_project_management_v1.yaml'
authority = yaml.safe_load(authority_path.read_text(encoding='utf-8'))

assert authority['department']['id'] == 'project-management'
assert authority['department']['status'] == 'persona-contract-only'
assert authority['kpis']['required_personas'] == 3
assert authority['kpis']['supported_action_count'] == 0
assert len(authority['personas']) == 3
assert all(persona['coverage_status'] == 'persona-contract-only' for persona in authority['personas'].values())
assert all(not persona['supported_actions'] for persona in authority['personas'].values())
```

```
with TemporaryDirectory() as temp_dir:
    pack = build_department_pack('project-management', store=JsonStore(Path(temp_dir)))

assert pack['department']['id'] == 'project-management'
assert pack['summary']['persona_count'] == 3
assert pack['summary']['active_persona_count'] == 0
assert pack['summary']['registry_backed_persona_count'] == 0
assert pack['summary']['supported_action_count'] == 0
assert pack['summary']['pack_state'] == 'blocked'
assert {persona['coverage_status'] for persona in pack['personas']} == {'persona-contract-only'}
```
