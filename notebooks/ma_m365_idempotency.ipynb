# MA: M365 Idempotency (INV-CAIO-M365-002, Eq. 3)

**Lemma:** [LEMMA_L2.md](../docs/contracts/caio-m365/LEMMA_L2.md) — Same Idempotency-Key + same body ⇒ second response = first (cached).

**Math:** \(\mathtt{sameKey}(H_1, H_2) \wedge \mathtt{sameBody}(R_1, R_2) \wedge \mathtt{success}(S_1) \Rightarrow S_2 = S_1\).

**Producer:** `scripts/ci/verify_m365_idempotency.py` → `configs/generated/m365_idempotency_verification.json`.

```python
import subprocess, sys
from pathlib import Path

repo_root = Path.cwd().resolve()
while not (repo_root / "scripts" / "ci" / "verify_m365_idempotency.py").exists() and repo_root != repo_root.parent:
    repo_root = repo_root.parent
r = subprocess.run([sys.executable, str(repo_root / "scripts" / "ci" / "verify_m365_idempotency.py")], cwd=str(repo_root), capture_output=True, text=True)
print(r.stdout or "")
assert r.returncode == 0
```
