# MA: M365 Audit (INV-M365-AUDIT-001, Eq. 5)

**Lemma:** [LEMMA_L4.md](../docs/contracts/caio-m365/LEMMA_L4.md) — When audit enabled, one record per instruction execution.

**Math:** \(\mathtt{auditEnabled}() \wedge \text{execution completed} \Rightarrow \exists\) exactly one audit record.

**Producer:** `scripts/ci/verify_m365_audit.py` → `configs/generated/m365_audit_verification.json`.

```python
import subprocess, sys
from pathlib import Path

repo_root = Path.cwd().resolve()
while not (repo_root / "scripts" / "ci" / "verify_m365_audit.py").exists() and repo_root != repo_root.parent:
    repo_root = repo_root.parent
r = subprocess.run([sys.executable, str(repo_root / "scripts" / "ci" / "verify_m365_audit.py")], cwd=str(repo_root), capture_output=True, text=True)
print(r.stdout or "")
assert r.returncode == 0
```

```python
import json
with open(repo_root / "configs" / "generated" / "m365_audit_verification.json") as f:
    a = json.load(f)
print("audit_pass:", a.get("audit_pass"))
```
