# L12 Proof Projection — Approval Contract Alignment

This projected proof notebook binds `L12` to the governed standalone-M365 approval-path contract used by `C1A` and to the deterministic approval-item lookup fallback required by the live `C1C` approval write path.

```
sample_items = [
    {"id": "21", "fields": {"ApprovalId": "lookup-1", "Status": "pending"}},
    {"id": "22", "fields": {"ApprovalId": "lookup-2", "Status": "approved"}},
]


def projected_lookup(items, approval_id):
    for item in items:
        if (item.get("fields", {}).get("ApprovalId") or item.get("id")) == approval_id:
            return item
    return None


assert projected_lookup(sample_items, "lookup-2")["id"] == "22"
assert projected_lookup(sample_items, "missing") is None
```
