# L45 — Persona Task Queues and State v1

## Statement
A shared queue authority, append-only store, and deterministic projection engine keep dashboard and persona task state aligned while invalid transitions fail closed.


## Proof Outline
1. The authority fixes valid statuses and transitions.
2. Created tasks and updates share one underlying store.
3. Projection order is deterministic.
4. Persona state is derived from explicit precedence rules.
