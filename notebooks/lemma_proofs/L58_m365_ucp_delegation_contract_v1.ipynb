# L58 Proof Replay

This notebook replays the bounded UCP delegation contract proof on fixed repo state.

## Proof Focus

- Six ordered delegation phases
- Five governance rules
- Persona count alignment with authoritative registry

```python
from pathlib import Path
import yaml

repo_root = Path.cwd()
contract = yaml.safe_load((repo_root / 'registry' / 'ucp_delegation_contract_v1.yaml').read_text(encoding='utf-8'))
registry = yaml.safe_load((repo_root / 'registry' / 'persona_registry_v2.yaml').read_text(encoding='utf-8'))

assert len(contract['delegation_phases']) == 6
assert len(contract['governance_rules']) == 5

personas = registry['personas']
total = len(personas)
contract_only = sum(1 for p in personas.values() if p.get('coverage_status') == 'persona-contract-only')
registry_backed = sum(1 for p in personas.values() if p.get('coverage_status') == 'registry-backed')

assert contract['kpis']['supported_persona_count'] == total
assert contract['kpis']['contract_only_persona_count'] == contract_only
assert contract['kpis']['registry_backed_persona_count'] == registry_backed
```
