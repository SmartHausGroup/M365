# L20 — Service Provisioning Site Detection Alignment

## Purpose

Capture the lemma proof that existing service provisioning must prefer the group-root SharePoint site over guessed `/sites/{mail_nickname}` paths.


```python
def resolve_existing_service_site(group_root_site: dict | None, path_site: dict | None) -> dict | None:
    if group_root_site and group_root_site.get("id"):
        return group_root_site
    if path_site and path_site.get("id"):
        return path_site
    return None

assert resolve_existing_service_site({"id": "site-1"}, None) == {"id": "site-1"}
assert resolve_existing_service_site(None, {"id": "site-2"}) == {"id": "site-2"}
assert resolve_existing_service_site(None, None) is None
```
