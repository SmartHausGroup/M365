# L55 Proof Replay

This notebook replays the bounded Studio Operations contract-pack proof on fixed repo state.

## Proof Focus

- One declared Studio Operations authority
- Five contract-only Studio Operations personas
- One deterministic zero-action claim
- One deterministic blocked-state rule


## Fixed-State Replay

Re-run the authority and runtime checks against the same repo state.

```
from pathlib import Path
from tempfile import TemporaryDirectory

import yaml

from smarthaus_common.department_pack import build_department_pack
from smarthaus_common.json_store import JsonStore

repo_root = Path.cwd()
authority_path = repo_root / 'registry' / 'department_pack_studio_operations_v1.yaml'
authority = yaml.safe_load(authority_path.read_text(encoding='utf-8'))

assert authority['department']['id'] == 'studio-operations'
assert authority['kpis']['required_personas'] == 5
assert authority['kpis']['supported_action_count'] == 0
```

```
with TemporaryDirectory() as temp_dir:
    pack = build_department_pack('studio-operations', store=JsonStore(Path(temp_dir)))

assert pack['department']['id'] == 'studio-operations'
assert pack['summary']['persona_count'] == 5
assert pack['summary']['active_persona_count'] == 0
assert pack['summary']['registry_backed_persona_count'] == 0
assert pack['summary']['supported_action_count'] == 0
assert pack['summary']['pack_state'] == 'blocked'
assert [persona['persona_id'] for persona in pack['personas']] == [
    'analytics-reporter',
    'finance-tracker',
    'infrastructure-maintainer',
    'legal-compliance-checker',
    'support-responder',
]
```
